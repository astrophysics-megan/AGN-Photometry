```python
import numpy as np
from astropy.io import fits
from astropy.table import Table
import matplotlib.pyplot as plt
import os
```

```python
cdir = os.getcwd()
obs_id = os.path.basename(os.path.dirname(os.path.dirname(cdir)))
mags = []
radii = [5.000, 7.000, 10.000, 12.000, 15.000, 18.000, 20.000, 25.000, 30.000, 35.000]
for radius in radii:
    maghist5 = fits.open("maghist_um2_"+str(radius)+"00_"+obs_id+".fits")
    mag_data = Table(maghist5[1].data)
    mags.append(mag_data['CORR_RATE'][0])
```

```python
plt.plot(radii, mags)
```
